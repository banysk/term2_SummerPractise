Импортируем необходимые библиотеки

```python
%matplotlib inline
import math
import scipy
import pandas as pd
from scipy import spatial
import numpy as np
from matplotlib import pyplot as plt
```

Задаём функцию f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

```python
def f(x):
    return math.sin(x/5.0) * math.exp(x/10.0) + 5 * math.exp(-x / 2.0)
```

Задаём список значений x в интервале от 1 до 16 с шагом 0.1
Вычисляем значение функции в этих точках

```python
x = np.arange(1, 16, 0.1)
y = map(f, x)
```

Строим график функции f(x)

```python
plt.plot(x, y)
```

Задаём функцию для создания системы линейных уравнений

```python
def get_matrix(x_set):
    n = len(x_set)
    matrix = []
    for x in x_set:
        row = []
        for i in range(n):
            row.append(x ** i)
        matrix.append(row)
    return np.array(matrix)
```

Задаём функцию для расчёта приближенных значений

```python
def get_approximate_values(w, x_set):
    n = len(w)
    y = []
    for x in x_set:
        y_curr = 0
        for i in range(n):
            y_curr += w[i]*(x ** i)
        y.append(y_curr)
    return y
```

Задаём функцию для составления и отрисовки формулы многочленов. Примечание: при x может стоять коэффициент == 0 из-за того, что значение крайне мало.

```python
def draw_formula(w_set):
    n = len(w_set)
    formula = ''
    for i in range(n):
        if i > 0 and w_set[i] > 0:
            formula += '+'
        formula += str(round(w_set[i], 2))
        if i == 0:
            continue
        formula += 'x^' + str(i)
        formula += ' '
    ax = plt.axes([0,0,0.1,0.1])
    ax.set_xticks([])
    ax.set_yticks([])
    ax.axis('off')
    plt.text(0.4,0.4,'$%s$' %formula,size=25,color="black")
```

Находим функцию f1(x) через 2 точки: 1, 15

```python
x_set_2 = [1, 15]
A_2 = get_matrix(x_set_2)
y_2 = np.array(map(f, x_set_2))
w_2 = scipy.linalg.solve(A_2, y_2)
y_approx_2 = get_approximate_values(w_2, x)
```

Строим график функции f(x) и f1(x)

```python
plt.plot(x, y)
plt.plot(x, y_approx_2)
```

```python
draw_formula(w_2)
```
